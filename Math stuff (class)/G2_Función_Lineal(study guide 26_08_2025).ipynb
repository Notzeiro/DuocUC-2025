# Problema 1

Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |
|$$ 500 $$                         | $$10 $$                  |
|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


```python
# 1. Defina variable dependiente e independiente, indicando unidad de medida.
print("La variable independiente es la cantidad de elementos medido en unidades, ya que esta es la\nentrada\
 y no depende de nada, y la variable dependiente es el tiempo de ejecucion medido en milisegundos,\nya que este dependera de\
 cuantos elementos entren a la funcion.\n")
# 2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo,
#  para esto utiliza la función *polyfit* de *Numpy*.
import numpy as np
from numpy import polyfit, array, poly1d , e

x=np.array([100 , 200 , 500 , 1000 , 2000])
y=np.array([2 , 4 , 10 , 20 , 40])

pendiente, intercepto = np.polyfit(x , y , 1)
#print(results)


def mongolo(x):
    return 0.02 * x + 0


print(f"La pendiente de la funcion es {pendiente:.4f} y el coeficiente de posicion (intercepto) es {intercepto:.4f}")
print(f"La funcion que relaciona el tamaño de la entrada con el tiempo de ejecucion es la siguiente:{mongolo}")

# 3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.

print(f"El tiempo de ejecucion para une entrada de 1.500 elementos es de: {round(mongolo(1500) , 2)} ms")

# 4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que
#  el tiempo de ejecución sea de $50$ milisegundos.

from scipy.optimize import fsolve

def g(x):
    return mongolo(x)-50

print(f"el tamaño de entrada que haria que el tiempo de ejecucion sea 50ms es:{fsolve(g , 10)} unidades")

```

4. ¿Cuánto tiempo toma una transferencia de $73,2$ GB?
5. Utiliza la función *fsolve* de *Scipy* para determinar la cantidad de datos que se puedan transferir en $123,5$ minutos.

```python
# 1. Defina variable dependiente e independiente, indicando unidad de medida.
print("La variable independiente son los datos transferidos (GB) y el tiempo de " \
"\ntransferencia en (minutos) es la variable dependiente.")
```

```python
# 2. Utilizando la función *polyfit* de *Numpy* encuentra la función que modela el tiempo 
# de transferencia $T(x)$ en función del volumen de los datos transferidos.
import numpy as np
from numpy import polyfit

x = np.array([5 , 10 , 25 , 50 , 100]) #variable independiente
y = np.array([10 , 20 , 50 , 100 , 200]) #variable dependiente

coeficientes= np.polyfit(x , y , 1)
print(coeficientes)

print(np.round(coeficientes , 2))

#funcion: ax + b // 0.02*x + 0
# es decir: T(x)=0.02x

print("La funcion es: T(x) = 2x + 0")
```

```python
# 3. Interprete la pendiente en el contexto del problema.

print("La pendiente es el coeficiente de 'X' dada la funcion (T(x)=2x + 0) la pendiente es: 2")
```

```python
# 4. ¿Cuánto tiempo toma una transferencia de $73,2$ GB?

#para responder esta pregunta debemos crear la funcion primero, sabemos que la funcion matematica es T(x)=0.02x

def t(x):
    return x*2

print(f"Una transferencia de 73,2GB tomara {round(t(73.2) , 2)} min en completarse")
```

```python
# 5. Utiliza la función *fsolve* de *Scipy* para determinar la cantidad de datos que se puedan transferir
#  en $123,5$ minutos.

from scipy.optimize import fsolve

tiempo_obj = 123.5

def f(x):
    return x*2 - tiempo_obj

solucion = fsolve(f , 50)

print(f"La cantidad de datos que se pueden transferir en 123,5 minutos es de: {solucion} GB")
```

```python
# 6. Grafica la función $T(x)$ indicando el nombre de los ejes y su unidad de medida.
# Interpreta lo observado. *Para graficar utiliza la biblioteca Matplotlib.*

import matplotlib.pyplot as plt

x_values = np.array([5 , 10 , 25 , 50 , 100])
y_values = np.array([10 , 20 , 50 , 100 , 200])

plt.plot(x_values , y_values)
plt.grid(True)


plt.title("T(x)= 2x + 0")
plt.xlabel("Cantidad de datos en GB")
plt.ylabel("Tiempo en minutos") 

plt.show()



```

# Problema 3


Una empresa de tecnología quiere optimizar el tiempo de carga y descarga de archivos en su servicio de almacenamiento en la nube. Los tiempos medios para diferentes tamaños de archivos son los siguientes:

<p><center>

| Tamaño del archivo (MB) | Tiempo de carga (s) | Tiempo de descarga (s)|
|-------------------------|---------------------|-----------------------|
|$$50$$                   |$$10$$               |$$8$$                  |
|$$100$$                  |$$20$$               |$$16$$                 |
|$$250$$                  |$$50$$               |$$40$$                 |
|$$500$$                  |$$100$$              |$$80$$                 |
|$$1.000$$                 |$$200$$              |$$160$$                |

</center></p>

1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$. Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).
2. Define variable dependiente e independiente, para ambas funciones, indicando su unidad de medida.
3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño del archivo si el tiempo de carga es de $163$ segundos.
5. Un trabajador insiste en que si el tiempo de descarga es de $195$ segundos, el tamaño del archivo es $1.200$ MB. ¿Tiene razón? Utiliza *Python* para determinar.
6. Grafica ambas funciones en el mismo gráfico, indicando nombre de los ejes y unidades de medida. Utiliza la librería *Matplotlib* para graficar.

```python
#1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$.
# Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).

import numpy as np
from numpy import polyfit , round , array

x = np.array([50 , 100 , 250 , 500 , 1000])
y = np.array([10 , 20 , 50 , 100 , 200])
y2 = np.array([8 , 16 , 40 , 80 , 160])

coeficientes = np.polyfit(x , y , 1)
coeficientes2 = np.polyfit(x , y2 , 1)

print("coeficientes 1 =" , np.round(coeficientes , 2))
print("coeficientes 2 =" , np.round(coeficientes2 , 2))

#la funcion al ser de grado 1 es de forma f(x) = ax + b 
#con los coeficientes podemos armar las funciones:

print("La funcion para el tiempo de carga es f(x)= 0.2x + 0 y para el tiempo de descarga: g(x)= 0.16x + 0")



```

```python
#2. Define variable dependiente e independiente, para ambas funciones, indicando su unidad de medida.

print("La variable independiente es el tamaño del archivo medido en MegaBytes y las variables dependientes" \
" serian los tiempos de carga y descarga, ambos medidos en segundos")
```

```python
#3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.

def carga(x):
    return 0.2*x

def descarga(x):
    return 0.16*x

funciones=[carga , descarga]

print("Para un archivo de 750 MB el tiempo de carga y descarga sera de:")
for funcion in funciones:
    print(f"{funcion(750)} segundos")
    
```

```python
#4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño del archivo si el tiempo de carga es de $163$ segundos.


from scipy.optimize import fsolve

def carga(x):
    return 0.2*x - 163

solucionn = fsolve(carga , 0)

print(f"el tamaño del archivo para un tiempo de carga de 163 segundos seria de: {solucionn} MB")
```

```python
#5. Un trabajador insiste en que si el tiempo de descarga es de $195$ segundos, el tamaño del archivo es $1.200$ MB. ¿Tiene razón?
# Utiliza *Python* para determinar.

def descarga(x):
    return 0.16*x

print(descarga(1200))

print("El tiempo de descarga para un archivo de 1200MB es de 192 segundos por lo que el trabajador no tiene razon")
```

```python
#6. Grafica ambas funciones en el mismo gráfico, indicando nombre de los ejes y unidades de medida.
# Utiliza la librería *Matplotlib* para graficar.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([50 , 100 , 250 , 500 , 1000])
y = np.array([10 , 20 , 50 , 100 , 200])
y2 = np.array([8 , 16 , 40 , 80 , 160])

plt.plot(x , y)
plt.grid(True)


plt.title("Carga")
plt.xlabel("Tamaño del archivo en MB")
plt.ylabel("Tiempo en segundos") 

plt.show()

plt.plot(x , y2)
plt.grid(True)


plt.title("Descarga")
plt.xlabel("Tamaño del archivo en MB")
plt.ylabel("Tiempo en segundos") 

plt.show()


```
